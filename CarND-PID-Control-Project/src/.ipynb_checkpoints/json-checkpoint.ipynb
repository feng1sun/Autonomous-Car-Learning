```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('./build/log.json') as f:
    data = json.load(f)
```

```python
cte = np.array(data['cte'])
d_error = np.array(data['d_error'])
i_error = np.array(data['i_error'])
p_error = np.array(data['p_error'])
u = np.array(data['control'])
```

```python
plt.plot(cte)
```

```python
plt.plot(d_error, label='d_error')
plt.plot(i_error, label='i_error')
plt.plot(p_error, label='p_error')
plt.legend()
```

```python
plt.plot(u)
```
