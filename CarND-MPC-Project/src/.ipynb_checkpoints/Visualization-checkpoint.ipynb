```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('../build/log.json') as f:
    data = json.load(f)
```

```python
data.keys()
```

```python
ptsx = np.array(data['ptsx'])
ptsy = np.array(data['ptsy'])
px = np.array(data['px'])
py = np.array(data['py'])
a = np.array(data['a'])
delta = np.array(data['delta'])
psi = np.array(data['psi'])
v = np.array(data['v'])

print('ptsx: ', ptsx.shape)
print('ptsx: ', ptsy.shape)
print('px: ', px.shape)
print('py: ', py.shape)
print('a: ', a.shape)
print('v: ', v.shape)
print('psi: ', psi.shape)
print('delta: ', delta.shape)
```

```python
plt.plot(px, py, 'r')
for x, y in zip(ptsx, ptsy):
    plt.plot(x, y)
```
